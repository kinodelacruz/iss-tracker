```python
#generating the data points
import requests
import json
import time

duration = 17*60
interval = 10

end = time.time()+duration

latitude = []
longitude = []
time_list = []

while time.time()<end:
    
    req = requests.get("http://api.open-notify.org/iss-now.json")
    data = json.loads(req.text)
    lat = float(data['iss_position']['latitude'])
    long = float(data['iss_position']['longitude'])
    timestamp = data['timestamp']
    latitude.append(lat)
    longitude.append(long)
    time_list.append(timestamp)
    print("Timestamp:", timestamp)
    print("Current latitude:", lat)
    print("Current longitude:", long)
    
    time.sleep(interval)
```

```python
#creating the graph

import matplotlib.pyplot as plt

plt.scatter(latitude, longitude)

plt.title('Tracking the ISS')
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```
